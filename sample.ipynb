```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
# set the default session seaborn style for graphics
sb.set() 
```

```python
pkmdf = pd.read_csv('Pokemon.csv')
pkmdf.head(5)
```

```python
# print the type and shape attributes
print("Data type : ", type(pkmdf))
print("Data dims : ", pkmdf.shape)
```

```python
print(pkmdf.dtypes)
```

```python
# extract a particular column from the dataframe
hp = pd.DataFrame(pkmdf['HP'])
print("Data type : ", type(hp))
print("Data dims : ", hp.size)
hp.head()
```

```python
# visualize using boxplot
sb.boxplot(data=hp, orient = "h")
```

```python
# visualise using hisplot
sb.histplot(hp, kde=True)
```

```python
# Visualize using violin plot
sb.violinplot(data=hp,orient = "h")
```

```python
attack = pd.DataFrame(pkmdf['Attack'])
```

```python
# set up matplot lib figure with three subplot
f, axes = plt.subplots(2, 3, figsize=(24, 12))

# plot the basic uni-variate figure for HP
sb.boxplot(data=hp, orient = "h", ax = axes[0,0])
sb.histplot(data=hp, kde=True, ax = axes[0,1])
sb.violinplot(data=hp,orient = "h", ax = axes[0,2])

# plot the basic uni-variate figure for attack

sb.boxplot(data=attack, orient = "h", ax = axes[1,0])
sb.histplot(data=attack, kde=True, ax = axes[1,1])
sb.violinplot(data=attack,orient = "h", ax = axes[1,2])
```
